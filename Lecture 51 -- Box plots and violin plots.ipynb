```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1 = randn(100)
```

```python
data2 = randn(100)
```

```python
sns.boxplot([data1,data2],orient='v')
```

```python
# violin plot

# normal dist

data1= stats.norm(0,5).rvs(100)

# 2 gamma distributions

data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       stats.gamma(5).rvs(50)*(-1)])
```

```python
sns.boxplot(data1,whis=np.inf)
sns.boxplot(data2)
```

```python
sns.violinplot(data1)
```

```python
sns.violinplot(data2,inner='stick')
```

```python
sns.violinplot(data2,bw=.01)
```
